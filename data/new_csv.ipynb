```python
import pandas as pd


Ratings = pd.read_csv(r'C:\Users\Math34\Downloads\archive\Ratings.csv')
Ratings = Ratings.sort_values(by='User-ID').head(100)


rating_users = Ratings['User-ID'].tolist()

rating_ISBN = Ratings['ISBN'].tolist()


Ratings.to_csv(r'C:\Users\Math34\Mathieu\Projets Ipssi\Ipssi_SQL_1\Ratings.csv', index=False)
```

```python
Users = pd.read_csv(r'C:\Users\Math34\Downloads\archive\Users.csv')


filtered_users = Users[Users['User-ID'].isin(rating_users)]

filtered_users.to_csv(r'C:\Users\Math34\Mathieu\Projets Ipssi\Ipssi_SQL_1\Users.csv', index=False)

```

```python
Books = pd.read_csv(r'C:\Users\Math34\Downloads\archive\Books.csv')


filtered_books = Books[Books['ISBN'].isin(rating_ISBN)]

filtered_books.to_csv(r'C:\Users\Math34\Mathieu\Projets Ipssi\Ipssi_SQL_1\Books.csv', index=False)
```
